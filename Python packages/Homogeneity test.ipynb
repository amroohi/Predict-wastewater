## Homogeneity Test

```python
import numpy as np
import pandas as pd
import pyhomogeneity as hg
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Loading data
df = pd.read_excel("C:/Users/sh/Desktop/term 3 ut/Thesis and Paper/R/New folder/TWWTP_raw_data.xlsx")
df
```

```python
var = df.columns
# var = ['COD_in']
for col in var:
    data = df[col]
    pettitt_res = hg.pettitt_test(data, alpha=0.05)
    snht_res = hg.snht_test(data, sim=10000)
    buishand_res = hg.buishand_u_test(data)
    result = pettitt_res

    mn = data.index[0]
    mx = data.index[-1]
    
    loc = pd.to_datetime(result.cp)
    mu1 = result.avg.mu1
    mu2 = result.avg.mu2
    
    plt.figure(figsize=(8,4), dpi=200)
    plt.plot(data,
             color='gray',
             label="Observation",
             lw=1)
    plt.hlines(mu1,
               xmin=mn,
               xmax=result.cp,
               linestyles='--',
               colors='k',
               lw=2,
               label='Mean 1 : ' + str(round(mu1,1)) + ' ')
    plt.hlines(mu2,
               xmin=result.cp,
               xmax=mx,
               linestyles='-.',
               colors='k',
               lw=2,
               label='Mean 2 : ' + str(round(mu2,1)) + ' ')
    plt.axvline(x=result.cp,
                linestyle=':',
                color='k',
                lw=2,
                label='Change point : '+ str(result.cp) + '\n p-value : ' + str(result.p))
    
#     plt.title('Homogeneity Test')
    plt.xlabel('Time(day)',
               fontsize = 13,
               x=0.5,
               y=-1.5)
    plt.ylabel(col,
               fontsize = 13,
               x=10,
               y=0.5)
    plt.legend(loc='best',
               fontsize = 7)
    
    # plt.savefig("F:/aaaaaa.jpg", dpi=600)
```

## Trend Test

```python
import numpy as np
import pandas as pd
import pymannkendall as mk
import matplotlib.pyplot as plt
import statsmodels.api as sm
%matplotlib inline

# Data generation for analysis
df = pd.read_excel("C:/Users/sh/Desktop/term 3 ut/Thesis and Paper/R/New folder/STWWTP_raw_data.xlsx")
df
```

```python
var = df.columns
var = ['MLSS_at']
for col in var:
    result = mk.original_test(df[col])
#     print(result.slope)
    fig, ax = plt.subplots(figsize=(16, 6))
    res = mk.original_test(df[col])
    trend_line = np.arange(len(df[col])) * res.slope + res.intercept
    
    ax.plot(df[col])
    ax.plot(df.index, trend_line)
    ax.legend(['data', 'trend line'])
    ax.set_xlabel('Day')
    ax.set_ylabel(col)
    ax.set_title('Trend Test')
```
